```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def reduce_resolution(image, block_size):
    # Create a kernel of the desired block size
    kernel = np.ones((block_size, block_size), dtype=np.float32) / (block_size * block_size)
    # Perform the convolution with the kernel
    reduced_resolution = cv2.filter2D(image, -1, kernel)
    return reduced_resolution

# Load image
image = cv2.imread("image.jpeg")

# Reduce resolution with a 3x3 block
reduced_3x3 = reduce_resolution(image, 3)

# Reduce resolution with a 5x5 block
reduced_5x5 = reduce_resolution(image, 5)

# Reduce resolution with a 7x7 block
reduced_7x7 = reduce_resolution(image, 7)

# Save the output images
cv2.imwrite("reduced_3x3.jpg", reduced_3x3)
cv2.imwrite("reduced_5x5.jpg", reduced_5x5)
cv2.imwrite("reduced_7x7.jpg", reduced_7x7)

# Display the images
plt.imshow(image)
plt.show()
plt.imshow(reduced_3x3)
plt.show()
plt.imshow(reduced_5x5)
plt.show()
plt.imshow(reduced_7x7)
plt.show()
```
