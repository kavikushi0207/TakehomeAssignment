```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load image
image = cv2.imread("image.jpeg")

# Get image dimensions
(h, w) = image.shape[:2]

# Get the rotation matrix for 45 degrees
M_45 = cv2.getRotationMatrix2D((w / 2, h / 2), 45, 1.0)

# Rotate the image by 45 degrees
rotated_45 = cv2.warpAffine(image, M_45, (w, h))

# Get the rotation matrix for 90 degrees
M_90 = cv2.getRotationMatrix2D((w / 2, h / 2), 90, 1.0)

# Rotate the image by 90 degrees
rotated_90 = cv2.warpAffine(image, M_90, (w, h))

# save the rotated images
cv2.imwrite("rotated_45.jpg", rotated_45)
cv2.imwrite("rotated_90.jpg", rotated_90)

# Display the images
plt.imshow(rotated_45)
plt.show()
plt.imshow(rotated_90)
plt.show()
```
