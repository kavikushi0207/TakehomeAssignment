```python
import cv2

def reduce_intensity_levels(image, levels):
    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Calculate the divisor for the intensity level reduction
    divisor = 256 / levels
    
    # Perform intensity level reduction
    reduced = gray // divisor * divisor
    
    return reduced

# Example usage
image = cv2.imread("image.jpeg")
reduced_image = reduce_intensity_levels(image, 2)
cv2.imwrite("reduced_image.jpeg", reduced_image)
```
